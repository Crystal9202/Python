# Random choose three characters from the list and make these three characters as a string

```python
import random

my_list = ['a','b','c','d','e','f','g','h','i','j']
result = random.sample(my_list, 3)
print(''.join(result))
```

# Write a roll the dice program as following. User can input if roll again or not(y/n)

```python
import random
min_num = 1
max_num = 6

roll_again = "yes"

while roll_again == "yes" or roll_again == "y":
    print("Rolling the dices...")
    print("The values are....")
    print(random.randint(min_num, max_num))
    print(random.randint(min_num, max_num))

    roll_again = input("Roll the dices again?")
```

# Please reference the following image to use random number to approximate pi  



```python
import random
from math import sqrt

inside=0
n=10000
for i in range(0,n):
    x=random.random()
    y=random.random()
    if sqrt(x*x+y*y)<=1:
        inside+=1
pi=4*inside/n
print(pi)
```

# Please use python to create a directory called "my_document" and create 26 file(a.txt, b.txt, c.txt, ......, z.txt)
